```python
#1. Créez une page contenant un  ́elément SVG déclaré de façon statique (pas de code JavaScript) d ́ecrivant un histogramme avec une barre pour
#chacune des données de l’exemple précédent. La surface de la barre doit être proportionnelle au champ number.

# Données d'exemple
data = [
    {"name": "A", "number": 400},
    {"name": "B", "number": 350},
    {"name": "C", "number": 300},
    {"name": "D", "number": 400},
    {"name": "E", "number": 100},
    {"name": "F", "number": 150},
    {"name": "G", "number": 400},
    {"name": "H", "number": 100}
]

# Ouvrir un fichier HTML en mode écriture
with open('histogramme_svg.html', 'w') as f:
    f.write('<html>\n')
    f.write('<body>\n')
    f.write('<svg width="400" height="250">\n')

    # Paramètres de l'histogramme
    bar_width = 40
    bar_padding = 10
    max_height = 200
    x_pos = 50
    y_pos = 50

    # Création des barres
    current_x = x_pos
    for d in data:
        bar_height = (d["number"] / max([item["number"] for item in data])) * max_height
        f.write(f'<rect x="{current_x}" y="{y_pos + max_height - bar_height}" width="{bar_width}" height="{bar_height}" fill="blue" />\n')
        current_x += bar_width + bar_padding

    f.write('</svg>\n')
    f.write('</body>\n')
    f.write('</html>\n')

```

```python
#2. Créez ce même histogramme de façon dynamique avec la bibliothèque
#d3.js. Chaque rectangle d ́ecrivant une barre aura un champ class dont
#la valeur sera C1, C2 ou C3 selon les cas. Des couleurs différentes seront attribuées à chacune des classes

# Données d'exemple
data = [
    {"name": "A", "number": 200},
    {"name": "B", "number": 150},
    {"name": "C", "number": 250},
    {"name": "D", "number": 100},
    {"name": "E", "number": 120},
    {"name": "F", "number": 80},
    {"name": "G", "number": 200}
]

# Paramètres de l'histogramme
bar_width = 40
bar_padding = 10
max_height = 200
x_pos = 50
y_pos = 50

# Génération du code HTML avec les données dynamiques
html_content = f'''
<!DOCTYPE html>
<html>
<head>
    <title>Histogramme avec d3.js</title>
    <script src="https://d3js.org/d3.v7.min.js"></script>
</head>
<body>

<script>

var data = {data};

var svg = d3.select("body")
            .append("svg")
            .attr("width", 400)
            .attr("height", 250);

svg.selectAll("rect")
    .data(data)
    .enter()
    .append("rect")
    .attr("x", function(d, i) {{ return {x_pos} + i * ({bar_width} + {bar_padding}); }})
    .attr("y", function(d) {{ return {y_pos} + {max_height} - (d.number / d3.max(data, function(d) {{ return d.number; }})) * {max_height}; }})
    .attr("width", {bar_width})
    .attr("height", function(d) {{ return (d.number / d3.max(data, function(d) {{ return d.number; }})) * {max_height}; }})
    .attr("fill", function(d, i) {{ 
        if (d.number == 250 || d.number == 200) {{
            return "red"; // Classe C1
        }} else if (d.number == 150 || d.number == 120) {{
            return "blue"; // Classe C2
        }} else {{
            return "green"; // Classe C3
        }}
    }})
    .attr("class", function(d, i) {{ 
        if (d.number == 250 || d.number == 200) {{
            return "C1";
        }} else if (d.number == 150 || d.number == 120) {{
            return "C2";
        }} else {{
            return "C3";
        }}
    }});

</script>

</body>
</html>
'''

# Écriture du code HTML dans un fichier
with open('histogramme_d3.html', 'w') as f:
    f.write(html_content)
```

```python
#On souhaite que chacune des données soit dor ́enavant d ́ecrite par une forme dont la position est définie par les coordonnées x et y. La surface
#de la forme est proportionnelle à l’attribut number, la couleur de remplissage correspond àl’association précédente, l’opacité de ce remplissage est
#défini par l’attribut intensity de la donnée. Enfin, la forme doit être un cercle si le type de la donn ́ee est A et un carré si le type de la donnée est B.

from jinja2 import Template
# Données d'exemple
data = [
    {"type": "A","x": 100, "y": 80, "number": 200, "intensity":0.4},
    {"type": "B","x": 130, "y": 120, "number": 150, "intensity":0.6},
    {"type": "A","x": 150, "y": 150, "number": 250, "intensity":0.7},
    {"type": "B","x": 200, "y": 200, "number": 100, "intensity":0.3},
    {"type": "A","x": 300, "y": 400, "number": 120, "intensity":0.7},
    {"type": "B","x": 200, "y": 350, "number": 80,  "intensity":0.4},
    {"type": "A","x": 350, "y": 100, "number": 200, "intensity":0.8}
]


# Génération du code HTML avec les données dynamiques
html_content = Template('''
<!DOCTYPE html>
<html>
<head>
    <title>Histogramme avec d3.js</title>
    <script src="https://d3js.org/d3.v7.min.js"></script>
    <style>
       .C1 {
         fill: red;
       }
       .C2 {
         fill: blue;
       }
       .C3 {
         fill: green;
       }
    </style>
</head>
<body>

<script>
var data = {{data | tojson}};

var svg = d3.select("body")
            .append("svg")
            .attr("width", 900)
            .attr("height", 900);

svg.selectAll("shape")
    .data(data)
    .enter()
    .append("g")
    .each(function(d) {
        if (d.type === "A") {
            d3.select(this).append("circle")
                .attr("cx", d.x)
                .attr("cy", d.y)
                .attr("r", Math.sqrt(d.number) * 5)
                .attr("class", function(d, i) {
                    if (d.number == 250 || d.number == 200) {
                        return "C1";
                    } 
                    else if (d.number == 150 || d.number == 120) {
                        return "C2";
                    }
                    else {
                        return "C3";
                    }
                })
                .style("opacity", d.intensity);
        }
        else if (d.type === "B") {
            d3.select(this).append("rect")
                .attr("x", d.x - Math.sqrt(d.number) * 5)
                .attr("y", d.y - Math.sqrt(d.number) * 5)
                .attr("width", Math.sqrt(d.number) * 10)
                .attr("height", Math.sqrt(d.number) * 10)
                .attr("class", function(d, i) {
                    if (d.number == 250 || d.number == 200) {
                        return "C1";
                    }
                    else if (d.number == 150 || d.number == 120) {
                        return "C2";
                    }
                    else {
                        return "C3";
                    }
                })
                .style("opacity", d.intensity);
        }
    });
    

</script>

</body>
</html>
''')

# Rendu du template avec les données
rendered_template = html_content.render(data=data)

# Écriture du code HTML dans un fichier
with open('histogramme_d3js_css.html', 'w') as f:
    f.write(rendered_template)
```
